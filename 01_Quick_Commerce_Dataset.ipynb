```python
#연습 문제 10. 사용자 정의 함수와 .apply() (단일 열)
#목표: 복잡한 조건문을 가진 함수를 만들어 특정 열의 데이터를 변환합니다.

#문제: Delivery_Time_Min(배달 소요 시간)이 30분 미만이면 'Fast',
#30분 이상이면 'Slow'를 반환하는 함수를 정의하십시오.
#이후 해당 함수를 Delivery_Time_Min 열에 적용하여, 속도(Fast/Slow)별 총 주문 건수(.size())를 집계하십시오.
```

```python
def time_categorise(time):
  if time < 30:
    return "Fast"
  else:
    return "Slow"
```

```python
#연습 문제 19. 다중 열의 값을 참조하는 사용자 정의 함수 (axis=1)
#목표: 한 행(가로줄)에 있는 여러 열의 데이터를 동시에 평가하여 새로운 범주를 부여합니다.

#문제: 배달 평가 등급을 산출하는 evaluate_delivery(row) 함수를 작성하십시오.
#     Delivery_Time_Min이 30분 미만이고 동시에 Delivery_Partner_Rating이 4점 이상이면 'Excellent',
#     그 외의 경우는 'Standard'를 반환해야 합니다.
      #이후 .apply()를 사용하여 전체 데이터프레임에 적용하고, 그 결과를 Delivery_Grade라는 새로운 열에 저장하십시오.
```

```python
def evaluate_delivery(row):
  if row["Delivery_Time_Min"] < 30 and row["Delivery_Partner_Rating"] >= 4:
    return "Excellent"
  else:
    return "Standard"
```

```python
df["Delivery_Grade"] = df.apply(evaluate_delivery, axis=1)
df.head()
```

```python
#연습 문제 20. 단일 열의 데이터 형변환 함수
#목표: 수치형 데이터를 특정 형식의 문자열로 변환하는 함수를 단일 열에 적용합니다.

#문제: 킬로미터 단위인 Distance_Km의 값을 미터(m) 단위로 변환하고,
#뒤에 'm' 문자열을 붙이는 km_to_meters(km) 함수를 작성하십시오. (예: 1.5 -> '1500m') 이를 Distance_Km 열에 적용하여 결과를 확인하십시오.
```

```python
def km_to_meters(Km):
  return f"{int(Km) * 1000}m"

km_to_meters(5)
```

```python
df.Distance_Km = df.Distance_Km.apply(km_to_meters)
df.head()
```

```python
#연습 문제 21. 그룹화 집계(.agg())에 사용자 정의 함수 적용
#목표: 내장된 통계 함수(mean, sum 등) 외에, 사용자가 직접 만든 계산 로직을 그룹 집계에 사용합니다.

#문제: 특정 열의 '최댓값과 최솟값의 차이(범위)'를 구하는 get_range(series) 함수를 작성하십시오.
#이후 Product_Category(상품 카테고리)별로 Order_Value(주문 금액)의 범위를 계산하십시오.
```

```python
def get_range(series):
  return series.max()-series.min()
```

```python
df.groupby("Product_Category")["Order_Value"].aggregate(get_range)
#사용자정의함수라면 aggregate() 따옴표가 없어야 함
```
